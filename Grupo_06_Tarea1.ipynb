**Pregunta 3:**

```python
datetime_index = pd.date_range(start='2023', periods=24, freq='M')
df_datetime =pd.DataFrame({'data': np.random.randint(100,501, size=24)}, index=datetime_index)
```

```python
df_datetime
```

**Pregunta 4:**

```python
df_datetime['lagged'] = df_datetime['data'].shift(periods=1)
```

```python
df_datetime['shifted'] = df_datetime['data'].shift(periods=-1)
```

```python
df_datetime['pct_change'] = df_datetime['data'].pct_change().mul(100)
```

```python
df_datetime['pct_change'].plot(title='Variación % de la variable macroeconómica')

plt.show()
```
