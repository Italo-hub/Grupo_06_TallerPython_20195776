```python
series = ['PN02528AQ', 'PN02529AQ', 'PN02533AQ', 'PN02530AQ']
start_date = '2002-01'
end_date = '2019-12'
freq = 'Trimestral'
```

```python
df= bcrp.bcrp_dataframe( series, start_date, end_date, freq ) 
df
```

```python
df['log'] = np.log(df['Producto bruto interno por tipo de gasto (millones S/ 2007) - Demanda Interna'])
df['diff'] = df['log'].diff().fillna(0)
df
```

```python
from statsmodels.tsa.stattools import adfuller

results = adfuller(df['log'])
print(results)
```

```python
print(' H0: Unit root', '\n', 'H1: No unit root')
print('The t-statistic of the test in Amazon is: ' + str(results[0]) + ' and a critical value of ' + str(results[4]['5%']))
print('The p-value of the test in Amazon is: ' + str(results[1]))
print('Since the calculated t-statistic does not exceed the critical value at 5% in absolute terms, we do not reject H0. There is a unit root')
print('Since the p-value is not close to zero, we do not reject H0. There is a unit root')
```

```python
from statsmodels.tsa.stattools import acf
from statsmodels.tsa.stattools import pacf
```

```python
fig = tsaplots.plot_pacf(df['diff'], lags = 25)
```

```python
fig_2 = tsaplots.plot_acf(df['diff'], lags= 25)
```

```python
#Podemos apreciar en ambos casos tanto en el ACF como en el PACF que no hay cambios bruscos o un punto de quiebre; por el
#contrario, estos parecen reflejar un descenso gradual, lo que nos llevaría a pensar que no es exactamente un modelo AR ni 
#un modelo MA, sino que se comporta más como si fuera un modelo ARMA.
```

```python
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
import numpy as np

model = ARIMA(df['diff'], order=(1,0,0))
results = model.fit()
residuo = results.resid
plt.figure(figsize=(12,6))
plt.plot(residuo)
plt.title('Modelo ARMA (1,0)')

plt.show()
```

```python
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
import numpy as np

model = ARIMA(df['diff'], order=(4,0,0))
results = model.fit()
residuo = results.resid
plt.figure(figsize=(12,6))
plt.plot(residuo)
plt.title('Modelo ARMA(4,0)')


plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_predict

model = ARIMA(df['diff'], order=(4,0,0))
results = model.fit()
start_forcasting = df['diff'].loc['2017-01-01':].index[0]
fig, ax = plt.subplots(figsize=(12,6))
df['diff'].plot(ax=ax, label='diff')
plot_predict(results, start=start_forcasting, end=len(df['diff']), ax=ax, plot_insample=False)
plt.title('Predicción modelo ARMA (4,0)')
plt.legend()
plt.show()
```
