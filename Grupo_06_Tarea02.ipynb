```python
def graph_time_series(axes, x, y, color, xlabel, ylabel):
  axes.plot(x, y, color=color)
  axes.set_xlabel(xlabel)
  axes.set_ylabel(ylabel, color=color)
  axes.tick_params(axis='y', colors=color)

def graph_time_series_max(axes, x, y, color, xlabel, ylabel):
  axes.plot(x, y, color=color)
  axes.set_xlabel(xlabel)
  axes.set_ylabel(ylabel, color=color)
  axes.tick_params(axis='y', colors=color)

  max_value = y.max()
  max_año = x[y.argmax()]
  axes.annotate('Punto Maximo',
                xy=(max_año, max_value),
                xytext=(max_año, max_value * 1.1),
                arrowprops=dict(facecolor='blue', arrowstyle='->'),
                fontsize=12, color='blue')
```

```python
plt.style.use('default')
fig, ax = plt.subplots(1,2, figsize=(18,9))
graph_time_series(ax[0], dato.index, dato['AAPL'], "darkblue", 'años','Precio de acciones Apple')
ax2=ax[0].twinx()
graph_time_series_max(ax2, dato.index, dato['NOK'],"darkred", 'años', 'Precio de acciones Nokia')

graph_time_series(ax[1], dato.index, dato['AMZN'], "darkblue", 'años','Precio de acciones Amazon')
ax2=ax[1].twinx()
graph_time_series_max(ax2, dato.index, dato['XOM'], "darkred", 'años','Precio de acciones de Exon Mobil Corporation')


fig.tight_layout(pad=1.0)
plt.show()
```
